## parsing the data collected by postman

```python
import json

class Item:
    def __init__(self, title, discount_percentage, final_price, image_link, description, product_link):
        self.title = title
        self.discount_percentage = discount_percentage
        self.final_price = final_price
        self.image_link = image_link
        self.description = description
        self.product_link = product_link

def read_items(json_str):
    try:
        json_obj = json.loads(fix_json_format(json_str))
        stations = json_obj["items"]
        items = []
        for item in stations:
            try:
                itm = save_item_data(item)
                items.append(itm)
            except KeyError as e:
                print(f"Missing key {e} in item: {item}")
        return items
    except json.JSONDecodeError as e:
        print(f"Failed to parse JSON: {e}")
        return []


def fix_json_format(json_str):
    remove_key = "\"mf_judgeme_widget\": ["
    while remove_key in json_str:
        x = json_str[:json_str.index(remove_key) - 1]
        json_str = json_str[json_str.index(remove_key):]
        json_str = json_str[json_str.index("multiple_product_type") - 2:]
        json_str = x + json_str
    return json_str


def save_item_data(item):
    title = item.get("title", "")
    discount_percentage = item.get("discount", [""])[0]
    final_price = item.get("price", [""])[0]
    image_link = item.get("image_url", "")
    description = item.get("description", "")
    product_link = item.get("product_url", "")
    return Item(title, discount_percentage, final_price, image_link, description, product_link)
```

```python
f = open('D://ukuleles.txt', 'r', encoding="utf8")
```

```python
file_contents = f.read()
```

```python
json_str = file_contents
items = read_items(json_str)
```

```python
d = []
n = 0
```

```python
for item in items:
    y = {'title': item.title,
        'discount_percentage': item.discount_percentage,
        'final_price': item.final_price,
        'image_link': item.image_link,
        'description': item.description,
        'product_link': item.product_link}
    d.append(y)
    n += 1
   # print(y)
```

## removing /n, /t, /w, whit spaces

```python
import json

# Load data from JSON file
with open('D:\\Aidetic\\Mehul\\Ukuleles.json', encoding="utf8") as f:
    products_data = json.load(f)
```

```python
import json

# Clean data by removing \n, \t, and \w
products_data_cleaned = []
for product in products_data:
    product_cleaned = {}
    for key, value in product.items():
        if isinstance(value, str):
            value = value.replace('\n', '').replace('\t', '').replace('\w', '')
        product_cleaned[str(key)] = str(value)
    products_data_cleaned.append(product_cleaned)
    
# Print cleaned data
print(products_data_cleaned)
```

```python
with open("Ukuleles.json", "w") as outfile:
    json.dump(products_data_cleaned, outfile)
    
# by using dump '' will conver in "" of dictinary key, pairs which will be valid for JSON
```
